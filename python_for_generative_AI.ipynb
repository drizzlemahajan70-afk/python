VARIABLE LENGTH ARGUMENT

1. NON-KEYWORD VARIABLE LENGTH ARGUMENT:-IT IS BEHAVE AS TUPLE.


REPRESENT BY STAR(*)



2.KEYWORD VARIABLE LENGTH ARGUMENT:-IT IS BEHAVE AS DICTIONARY.


REPRESENT BY DOUBKE STAR(**)



```python
#non keyword variable length argument
def test(*args):
  print(args)
  print(len(args))
  print(type(args))
```

```python
test(2,3,4,5,6)
```

```python
def sum_number(*args):
  sum = 0
  for num in args:
    sum+=num
  return sum
```

```python
sum_number(1,2,3,4,5,6)
```

```python
#keyword variable length argument
def test(**kwargs):
  print(kwargs)
  print(len(kwargs))
  print(type(kwargs))
```

```python
test(a=10,b=20,c=30)
```

```python
def pass_students(**kwargs):
  pass_student_list=[]
  for name, marks in kwargs.items():
    if marks>35:
      pass_student_list.append(name)
  return pass_student_list

```

```python
pass_students(A = 10, B = 20, C = 40, D = 50, E = 60, F = 70)
```

```python
cgp = np.array([6,6.5,7,7.8,8,8.6,9,9.2,9.5,10])
package = np.array([4,4.5,5,5.8,6.2,6.8,7,7.5,8,8.8])
font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}
plt.plot(cgp,package, "*:g", ms = 15, mec = "pink", mfc ="yellow",linewidth = 2.4)
plt.xlabel("CGP",fontdict = font1)
plt.ylabel("PACKAGE",fontdict = font2)
plt.title("CGP VS PACKAGE")
plt.show()
```
